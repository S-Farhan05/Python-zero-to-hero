```python
# List comprehension and for loops

Calcius_temp = [22,12,44,55.3]

fehrenhite_temp = [((9/5)*temp+32) for temp in Calcius_temp]
print(fehrenhite_temp)

newtemp = [11,23,1312,122]
fehra=[]
for i in newtemp:
    fehra.append(((9/5)*i)+32)
print(fehra)
```

```python
my_list6= [num**3 for num in range(1,16) if (num**3)%4==0]
print(my_list6)

my_list7=[]

for i in range(1,16):
    if (i**3)%4==0:
       
        my_list7.append(i**3)
       

print(my_list7)
```

```python
my_list8 = [str(num) for num in range(1,51) if num%3 == 0 or num%5==0]
print(my_list8)
print(type(my_list8))


my_list9 =[]

for num in range(1,51):
    if num%3 == 0 or num%5==0:
        my_list9.append(str(num))
       

print(my_list9)
```
