```python
def display_list(groceries):
    print(f"The Current List is : {groceries}")
```

```python
def add_new_item(groceries):
    state = True
    while state:
        val = input("What item do you want to add :? ")

        if val in groceries:
            print("The item already existed in the list")
            continue
        else:
            groceries.append(val)
            print(f"{val} Added Succesfully ")
            break
        
```

```python
def remove_list(groceries):
    state =True
    while state:
        display_list(groceries)
        delete_index =  input("What do u want to remove from this list (Index type)")

        try:
            int(delete_index)
        
        except:
            print("The value must be an valid integer")
            continue
        else:
            if  int(delete_index) < 0 or int(delete_index) > len(groceries) -1:
                print("Not a valid index , index start from 0")
            
            else:
                groceries.pop(int(delete_index))
                break
                
            
```
