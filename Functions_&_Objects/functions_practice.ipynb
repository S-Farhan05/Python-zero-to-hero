Tuple Unpacking with functions


```python
def employer_of_the_month(empolyer_info:list):

    current_max = 0 
    empolyer_name=''

    for name , hrs in empolyer_info:

        if hrs>current_max:
            current_max=hrs
            empolyer_name=name

    return current_max,empolyer_name
```

```python
hrs , name = employer_of_the_month([('Ayesh',180),('hamza',45),('Azan',200),('Faris',150)])
print(name ,f"Hours worked by empolyer: {hrs} ")
```
