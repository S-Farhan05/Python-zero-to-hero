```python
# --- Inheritence python

class Animal():

    def __init__(self,color):
        print("I am Animal class")
        self.color = color

    def eat(self):
        print("Animal eats")


class Dog(Animal):

    def __init__(self):
        print("I am a dog")



myAnimal = Animal('Blue')
myAnimal.eat()
dog = Dog()

print(myAnimal.color)
```

```python
#-- Polymorphism -- #

class Dog():

    def __init__(self,name):
        self.name=name
    
    def speak(self):
        print(self.name + " Barks")

class Cat():

    def __init__(self,name):
        self.name=name
    
    def speak(self):
        print(self.name + " Meows")


Carl = Dog(name="Carl")
Felix = Cat(name="Felix")
Carl.speak()
Felix.speak()
```

```python
for pet in [Carl,Felix]:
    print(type(pet))
    
```

```python
def pet_speak(pet):
    pet.speak()


pet_speak(Felix)
print(type(Felix))
print(type(Cat))
```

```python
# -- Abstract Classes -- #

class Animal():
    def __init__(self,name):
        self.name=name

    def speak(self):
        raise NotImplementedError("Abstarct method must be class for child class")
```
